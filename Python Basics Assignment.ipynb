```python
#6.How do conditional statements work in Python? Illustrate with examples.

Answer:-
conditional statement:- Conditionals in Python are used to execute code based on the evaluation of one or more conditions.

Type of conditional statement:
1. if
2. if-else
3. if-elif-else
4. nested if-else

Example of conditional statement:
1. if
a = 200 
if a > 100:
    print("the no is greater than 100")

Results- the no is greater than 100
    
    
2. if-else
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
    
Results- x is greater than 5


3. if-elif-else
a = 10
if a > 100:
    print("will execute if is a greater than 100")
elif a < 100:
    print("will execute if is a lesser than 100")
else:
    print("the no is equal to 100")
    
Results- will execute if is a lesser than 100


4. nested if-else
x = 10
y = 5

if x > 5:
    if y > 5:
        print("both x and y is greater than 5")
    else:
        print("x is greater than 5 but y is not")
else:
    print("x is grater than 5")
    
Results- x is greater than 5 but y is not
```
